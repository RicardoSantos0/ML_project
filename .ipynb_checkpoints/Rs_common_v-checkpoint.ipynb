```python
df_train[['Gender','Name']] = df_train.Name.str.split(" ", n=1, expand=True)

df_test[['Gender','Name']] = df_test.Name.str.split(" ", n=1, expand=True)
```

```python
df_train['Gender'].replace(['Miss', 'Mrs.', 'Mr.'], [0,0,1],inplace=True)
df_train = df_train.drop("Name", axis=1)

df_test['Gender'].replace(['Miss', 'Mrs.', 'Mr.'], [0,0,1],inplace=True)
df_test = df_test.drop("Name", axis=1)
```

```python
#replaces index by ID of citizens; drops previous
df_train = df_train.set_index('CITIZEN_ID', drop=True)

df_test = df_test.set_index('CITIZEN_ID', drop=True)
```

```python
#replaces ? with np.nan
df_train.replace('?', np.nan, inplace=True)

df_test.replace('?', np.nan, inplace=True)
```

```python
#calculates age and drops birthay column
birthday = df_train["Birthday"].str.split(",", n=1, expand=True).rename(columns={0:"Day and month", 1:"Year"})
df_train = df_train.assign(Birthday=birthday['Year'])
df_train["Birthday"] = df_train["Birthday"].astype(int)
df_train["Birthday"] = 2048 - df_train["Birthday"]
df_train.rename(columns={'Birthday':'Age'}, inplace=True)

birthday = df_test["Birthday"].str.split(",", n=1, expand=True).rename(columns={0:"Day and month", 1:"Year"})
df_test = df_test.assign(Birthday=birthday['Year'])
df_test["Birthday"] = df_test["Birthday"].astype(int)
df_test["Birthday"] = 2048 - df_test["Birthday"]
df_test.rename(columns={'Birthday':'Age'}, inplace=True)
```

```python
#merges money received and ticket price columns; people who payed have now a negative value in column "Money Received"
df_train["Money Received"] = -1 * df_train["Money Received"]
df_train['Ticket Price']= df_train['Money Received'] + df_train['Ticket Price']
del df_train['Money Received']

df_test["Money Received"] = -1 * df_test["Money Received"]
df_test['Ticket Price'] = df_test['Money Received'] + df_test['Ticket Price']
del df_test['Money Received']
```

```python
def bar_charts_categorical(df, feature, target):
    cont_tab = pd.crosstab(df[feature], df[target], margins = True)
    categories = cont_tab.index[:-1]
        
    fig = plt.figure(figsize=(15, 5))
    
    plt.subplot(121)
    p1 = plt.bar(categories, cont_tab.iloc[:-1, 0].values, 0.55, color="gray")
    p2 = plt.bar(categories, cont_tab.iloc[:-1, 1].values, 0.55, bottom=cont_tab.iloc[:-1, 0], color="skyblue")
    plt.legend((p2[0], p1[0]), ('$y_i=1$', '$y_i=0$'))
    plt.xticks(rotation='vertical')
    plt.title("Frequency bar chart")
    plt.xlabel(feature)
    plt.ylabel("$Frequency$")

    # auxiliary data for 122
    obs_pct = np.array([np.divide(cont_tab.iloc[:-1, 0].values, cont_tab.iloc[:-1, 2].values), 
                        np.divide(cont_tab.iloc[:-1, 1].values, cont_tab.iloc[:-1, 2].values)])
      
    plt.subplot(122)
    p1 = plt.bar(categories, obs_pct[0], 0.55, color="gray")
    p2 = plt.bar(categories, obs_pct[1], 0.55, bottom=obs_pct[0], color="skyblue")
    plt.xticks(rotation='vertical')
    plt.legend((p2[0], p1[0]), ('$y_i=1$', '$y_i=0$'))
    plt.title("Proportion bar chart")
    plt.xlabel(feature)
    plt.ylabel("$p$")

    plt.show()
```

### Training and Testing

We have to be wary of possible overfitting of the model. Overfitting, in essence is a process by which the ML model follows the sample data to an excessive degree. In theory this should be a good thing. However, overfitting means that the model adjusted so well to the data that, actually, it fits the data (and its outliers) so much, that patterns and trends become overshadowed.

In practical means, this would mean that the model would be an unreliable predictor. It is hard to spot so, a method usually used is to actually split the data into testing data and training data. The model will be trained with a portion of the data. Afterwards, the remaining portion is used to test the data.

If a model is overfitted, the predictions of the test would have low accuracy. This would mean that, although the model adjusted particularly well to the training data, it failed to grasp the underlying logic behind the data distribution.

```python
#create regression with the training data. This will be a log-lin regression.
from sklearn.model_selection import train_test_split
#we assing to each test and train variables. The method splits them automatically
#test size is 0.2 (20% of the data will be used to test, a random state allows us to make sure that the data distribution follows a particular pattern)
x_train, x_test, y_train, y_test = train_test_split(inputs_scaled, targets, test_size = 0.2, random_state = 42)
```

```python
sns.set()
from sklearn.linear_model import LinearRegression
```

```python
#create regression with the training data. This will be a log-lin regression.
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
y_hat = reg.predict(x_train)
```

```python
plt.scatter(y_train,y_hat)
plt.xlabel('Output Training Data', fontsize = 20)
plt.ylabel('Predicitions of output of the model', fontsize = 20)
plt.show()

#In general, we can see that the model tends to place our predictions near the real output values. The fit is not perfect, but accpetable.
```
